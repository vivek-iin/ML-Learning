```python
import pandas as pd

data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9], 'D': [10, 11, 12]}
df = pd.DataFrame(data)

print(df)
```

```python
df = df.iloc[:, 2:]
print(df)
```

* Summary: df = df.iloc[:, 2:]
* This removes the first two columns (A and B).
* The updated df contains only columns from index 2 onward (C and D).

* Breakdown:
* df.iloc[:, 2:] → Uses .iloc[], which is used for integer-location based indexing.
* : → Selects all rows.
* 2: → Selects columns starting from the 3rd column (index 2) onward, dropping the first two columns (index 0 and 1).
